```python
#while loops
count = 0
while count < 25:
    count+=1
    print(count)
#while loop runs until the condition is true.
#when the condition becomes false, it stops executing.
#In this case, it will print numbers from 1 to 25.
#the interaction process.
```

```python
num2 =100
while num2 > 0:
    num2-=2
    print(num2)
```
